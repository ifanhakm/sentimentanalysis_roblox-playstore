```python
from google_play_scraper import reviews, Sort
import pandas as pd
from google.colab import files
```

Take reviews from Roblox in Google Playstore for all stars rating and limit it to 3052 reviews based on newest reviews in Indonesia with English language

```python
scrapreview, continuation_token = reviews(
    "com.roblox.client",          # ID app
    lang="en",                    # defaults to ‚Äòen‚Äô
    country="id",                 # defaults to ‚Äòus‚Äô
    sort=Sort.NEWEST,             # defaults to Sort.MOST_RELEVANT
    count=3052,                   # jumlah ulasan yang diinginkan
)
```

```python
app_reviews_df = pd.DataFrame(scrapreview)
app_reviews_df.to_csv("/content/roblox_playstore.csv", index=None, header=True)
files.download("roblox_playstore.csv")
```
